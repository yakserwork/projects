# Исследование объявлений о продаже квартир

Нужно научиться определять рыночную стоимость объектов недвижимости. В нашем распоряжении архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Задача — установить параметры, это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.

 2. Предобработка данных.
 3. Расчеты и добавление результатов в таблицу.
 4. Исследовательский анализ данных.
 5. Общий вывод.

## Предобработка данных

В первую очередь приведем названия столбцов к единому стилю. Переименуем проблемные заголовки методом `rename` и проверим результат.

```python
df = df.rename(columns={ # переименование столбцов
    'cityCenters_nearest': 'city_centers_nearest',
    'open_plan': 'is_open_plan',
    'open_plan': 'is_open_plan',
    'parks_around3000': 'parks_around_3000',
    'ponds_around3000': 'ponds_around_3000',
    'studio': 'is_studio'}
)

df.columns.tolist() # проверка названий столбцов, приведение к списку
```

Рассмотрим аномалию в столбце `rooms`. Максимальное количество комнат - 19, выглядит неправдоподобно. Посмотрим уникальные значения.

```python
df['rooms'].value_counts() # поиск и подсчет уникальных значений столбца 'rooms'
```

Мы обнаружили 197 нулевых значений, эти ячейки можно удалить. Аномально большие значения обработаем позже на этапе изучения параметров.

```python
df = df[df['rooms'] != 0]
```

Для дальнейшего исследования переведем цены в млн. руб.

```python
df['last_price'] = df['last_price'] / 1000000 # перевод значений цены из руб. в млн. руб.
```

### Поиск пропущенных значений

Избавимся от пропусков. Найдем пропущенные значения в столбцах и посчитаем их количество.

```python
df.isna().sum() # поиск пропущенных значений пропусков в ячейках
```

Больше всего пропущенных значений в столбце `is_apartment`. Это булев тип и наиболее вероятно, что пропуски указывают на значение `False`.

```python
df['is_apartment'] = df['is_apartment'].fillna(False) # замена пропусков на значение False
```

Пропуски в столбцах `floors_total ` и `locality_name` составляют менее 1% датасета. Их можно удалить.

```python
df = df.dropna(subset=['floors_total', 'locality_name']) # удаление пропусков
df = df.reset_index(drop=True) # сброс индекса
```

Рассмотрим ситуацию с пропусками в столбцах:
* `parks_around_3000`
* `parks_nearest`
* `ponds_around_3000`
* `ponds_nearest`
* `airports_nearest`
* `city_centers_nearest`

Наиболее оптимальное решение - заполнить пропуски медианой в зависимости от населенного пункта. Напишем цикл и проверим результат.

```python
column_list = [               # список столбцов для цикла
    'parks_around_3000',
    'parks_nearest',
    'ponds_around_3000',
    'ponds_nearest',
    'airports_nearest',
    'city_centers_nearest',
]

for column in column_list: # цикл для замены пропусков медианным значением по категориям
    df[column] = df[column].fillna(df.groupby('locality_name')[column].transform("median"))

df.isna().sum() # поиск пропусков
```

К сожалению, не все пропуски удалось заполнить. Вероятно, по некоторым группам все значения пропущены. Оставим их без изменений.

Пропуски в столбце `balcony` могут означать, что в квартире нет балкона. Заменим пропуски нулями.

```python
df['balcony'] = df['balcony'].fillna(0) # замена пропусков нулями
```

Пропусков в столбце `ceiling_height` более 30% от всего датасета. Изучим подробнее значения стобца.

```python
df['ceiling_height'].describe() # разброс значений
```

Большая часть значений высоты близка к медиане. Так как это количественная переменная, то будет правильнее заполнить пропуски медианным значением. Для более точной замены пропусков, добавим зависимость от количества этажей в доме.

```python
df['ceiling_height'] = df['ceiling_height'].fillna(df.groupby('floors_total')['ceiling_height'].transform("median")) # замена пропусков медианным значением по категориям
```

При изучении значений также была обнаружена аномалия. Минимальная и максимальная высота потолка составляет 1 и 100 метров. Здесь явно ошибка, посмотрим сколько таких строк в датасете. Среднее значение примерно 3 метра, значит нормальная высота должна быть от 2 до 4 метров.

```python
len(df[(df['ceiling_height'] < 2) | # подсчет строк по двум условиям
   (df['ceiling_height'] > 4)])
```

Всего 79 строк, это менее 1% от датасета, поэтому можем их удалить с помощью среза. Проверим результат.

```python
df = df.query('2 < ceiling_height < 4') # применение среза по двум условиям
df['ceiling_height'].describe() # разброс значений
```

Пропуски удалены и сейчас разброс значений выглядит более правдоподобно. 

Рассмотрим пропуски в столбце `days_exposition`. Вероятно отсутствие значений означает, что на момент выгрузки данных, квартиры были еще в продаже. Пропуски не помешают исследованию, но возможно на этапе изучения параметров потребуется сделать срез. Чтобы не потерять строки, заменим пропуски на значение -1.

```python
df['days_exposition'] = df['days_exposition'].fillna(-1)
```

Пропуски в столбцах `living_area` и `kitchen_area` могут сильно повлиять на результат исследования, поэтому их нужно заполнить максимально точно. Проверим корреляцию и соотнесем значения жилой площади и площади кухни с общей площадью.

```python
print(f'Коэффициент корреляции между жилой площадью и общей площадью: {df["living_area"].corr(df["total_area"])}') # корреляция
print(f'Коэффициент корреляции между площадью кухни и общей площадью: {df["kitchen_area"].corr(df["total_area"])}')
```

Корреляция достаточно большая, чтобы заменить пропуски медианными значениями в зависимости от общей площади. Проверим результат.

```python
df['living_area'] = df['living_area'].fillna(df.groupby('total_area')['living_area'].transform("median")) # замена пропусков медианным 
df['kitchen_area'] = df['kitchen_area'].fillna(df.groupby('total_area')['kitchen_area'].transform("median")) # значением по категориям
df.isna().sum() # поиск пропусков
```

Осталось небольшое количество пропусков столбцах, 132 и 185. Для них не нашлось подходящих значений общей площади, поэтому удалим их.

```python
df = df.dropna(subset=['living_area', 'kitchen_area']) # замена пропусков нулями
df = df.reset_index(drop=True) # сброс индекса
```

В некоторых столбцах тип данных не соответствует содержимому. Для оптимизации поменяем тип данных в столбцах и проверим результат.

* `is_apartment` в булев тип, так как значение сообщает является ли апартаментами или нет
* `floors_total`, `balcony` - в целочисленные, так как не могут быть дробными
* `first_day_exposition` преобразуем во время

```python
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S') # перевод в формат времени
df = df.astype({'is_apartment': 'bool', 'floors_total': 'int64', 'balcony': 'int64'}) # изменения типа данных в столбцах
```

### Поиск дубликатов

Теперь проверим наличие явных дубликатов.

```python
df.duplicated().sum() # подсчет явных дубликатов
```

Явных дубликатов нет. Проверим неявные дубликаты в `locality_name`, возможны отличия в написании одного и того же населенного пункта.

```python
df['locality_name'].unique() # уникальные значения столбца
```

При изучении значений, обнаружено разное написание в типе населенного пункта (поселок, посёлок). Оставим одно значение `поселок`.

```python
df['locality_name'] = df['locality_name'].str.replace('посёлок', 'поселок') # замена значений
```

Проверим какое количество строк осталось после предобработки.

```python
print(f' Доля удаленных строк {round((rows_after - len(df)) / rows_after * 100)}%') # подсчет доли строк
```

### Вывод

Мы заполнили пропуски там, где это было возможно. Пропуски появились либо из-за того, что часть данных была вписана пользователями или из-за ошибки в работе картграфического сервиса. Заполнять пропуски нулями было бы необъективно. Также мы избавились от неявных дубликатов. 

Часть строк была удалена, но это не повлияет на результат исследования, так как доля не превышает 2%.

## Расчёты и добавление результатов в таблицу

Для проведения анализа нам потребуются следующие данные:

* Цена квадратного метра
* День недели, месяц и год публикации объявления
* Этаж квартиры (варианты — первый, последний, другой)
* Соотношение жилой и общей площади, а также отношение площади кухни к общей

Расчитаем эти данные и добавим в таблицу.

Добавим столбец с ценой квадратного метра. Для этого разделим стоимость квартиры на общую площадь.

```python
df['sqm_price'] = df['last_price'] / df['total_area'] # расчет цены квадратного метра
```

Добавим столбцы с днем недели, месяцем и годом. Для этого вызовем соответствующие методы.

```python
df['weekday'] = df['first_day_exposition'].dt.weekday # вызов дня недели
df['month'] = df['first_day_exposition'].dt.month # вызов месяца
df['year'] = df['first_day_exposition'].dt.year # вызов года
```

Добавим столбец с этажом квартиры. Для этого напишем функцию.

```python
def floor_type(row):              # фунция для категоризации этажей
    if row['floor'] == 1:
        return 'Первый'
    elif row['floor'] == row['floors_total']:
        return 'Последний'
    else:
        return 'Другой'
    
df['floors_cat'] = df.apply(floor_type, axis=1) # добавление столбца с категорией этажа
```

Добавим столбцы с соотношением жилой и общей площади и отношением площади кухни к общей. Выведем на экран для проверки.

```python
df['living_ratio'] = df['living_area'] / df['total_area'] # соотношение
df['kitchen_ratio'] = df['kitchen_area'] / df['total_area'] # площадей
display(df)
```

## Исследовательский анализ данных

### Изучение параметров

Рассмотрим следующие параметры: 

* площадь
* цена
* число комнат
* высота потолков

Так как мы будем рассматривать несколько значений, то предварительно напишем функцию для построения гистограммы и диаграммы размаха по каждому параметру, а также выведем на экран разброс значений с помощью метода describe.

```python
def visual_1(df, column, bins, a, b, c, d): # функция построения графиков

    plt.figure(figsize=(15, 5))
    
    plt.hist(df[column], bins=bins, range=(a, b)) # гисторграмма с аргументами
    plt.grid(axis = 'both') # включение сетки
    plt.show() # показать фигуру
    
    print(df[column].describe()) # диаграмма размаха
    
    plt.figure(figsize=(15, 5))
    
    plt.boxplot(df[column], vert=False)
    plt.grid(axis = 'both') # включение сетки
    plt.xlim(c, d) # размеры по горизонтали
    plt.show() # показать фигуры
```

### Сравнение населенных пунктов по количеству объявлений

Выберем 10 населенных пунктов с наибольшим числом объявлений. Посчитаем среднюю цену квадратного метра в этих населенных пунктах. Выделим населенные пункты с самой высокой и низкой стоимостью жилья.

```python
df_loc = df_new.pivot_table( # сводная таблица по нас. пункту
    index='locality_name',
    values='sqm_price',
    aggfunc=['count','mean']
)

df_loc.columns = ['count', 'mean_price'] # названия столбцов

most_listed_cities = df_loc.sort_values(by='count', ascending=False).head(10).reset_index(drop=False) # сортировка по количеству

print('10 населённых пунктов с наибольшим числом объявлений:')
display(most_listed_cities) # вывод отсортированной таблицы

print(f'Населённый пункт с самой высокой стоимостью жилья из топ-10: {most_listed_cities.loc[most_listed_cities["mean_price"] == most_listed_cities["mean_price"].max(), "locality_name"].item()}')    
print(f'Населённый пункт с самой низкой стоимостью жилья из топ-10: {most_listed_cities.loc[most_listed_cities["mean_price"] == most_listed_cities["mean_price"].min(), "locality_name"].item()}')
```

Больше всего объявлений в Санкт-Петербурге и здесь же самая высокая цена за квадратный метр. Меньше всего объявлений в Выборге и самая низкая цена за квадратный метр, в 2 раза меньше чем в Санкт-Петербурге.

Изучим данные для этой территории, построим гистограммы и диаграммы размаха для параметров: 

* площадь

* высота потолков

Для сравнения и наглядности наложим гистограммы друг на друга. Напишем функцию для наложения гистограмм.

```python
def comparsion(df_1, df_2, column, f, t, bins): # функция для наложения гистограмм
    ax = df_1.plot(kind='hist', y=column, range=(f, t), bins=bins, linewidth=5, alpha=0.7, label='all_data', figsize=(15, 5)) # гистограмма 1
    df_2.plot(kind='hist', y=column, range=(f, t), bins=bins, label='spb_center', ax=ax, grid=True, legend=True, figsize=(15, 5)) # гистограмма 2
    print(df_2[column].describe()) 
    np.warnings.filterwarnings('ignore') # игнорирование предупреждений
```

Построим гистограмму для площади.

```python
comparsion(df, df_spb_center, 'total_area', 0, 300, 30) # вызов функции
```

Большинство значений не превышают 100 кв. м. Выбросы с 200 кв. м. Можно сделать вывод, что в центре города продаются квартиры с большей площадью.

Построим гистограмму для цены.

```python
comparsion(df, df_spb_center, 'last_price', 0, 30, 50) # вызов функции
```

Большинство значений не превышают 15 млн. руб. Выбросы с 25 млн. руб. Из этого следует, что в центре квартиры значительно дороже.

Построим гистограмму для количества комнат.

```python
comparsion(df, df_spb_center, 'rooms', 0, 7, 7) # вызов функции
```

Можем заметить, что в центре больше предложений квартир с 2-3 комнатами, тогда как за пределами центра больше предложений с 1-2 комнатами.
